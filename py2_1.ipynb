```python
#stackplot
import pandas as pd
import matplotlib.pyplot as plt
data={'year':[2021,2022,2023,2024,2025],'sales':[100,120,150,170,200],'expenses':[70,80,90,100,170]}
df=pd.DataFrame(data)
plt.stackplot(df['year'],df['sales'],df['expenses'],labels=['sales','expenses'],colors=['blue','yellow'])
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={'year':[2021,2022,2023,2024,2025],'sales':[100,120,150,170,200],'expenses':[70,80,170,100,250]}
df=pd.DataFrame(data)
fig,ax=plt.subplots()
ax.set_xticks([2021,2022,2023,2024,2025])
ax.plot(df['year'],df['expenses'],color='green')
ax.plot(df['year'],df['sales'],color='red')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={'year':[2021,2022,2023,2024,2025],'sales':[100,120,150,170,200],'expenses':[70,80,170,100,250]}
df=pd.DataFrame(data)
fig,ax=plt.subplots()
ax.set_xticks([2021,2022,2023,2024,2025])
ax.plot(df['year'],df['expenses'],color='green')
ax.plot(df['year'],df['sales'],color='red')
ax.fill_between(df['year'],df['sales'],df['expenses'],where=df.sales >df.expenses,interpolate=True,color='red')
ax.fill_between(df['year'],df['sales'],df['expenses'],where=df.sales <df.expenses,interpolate=True,color='red')
plt.show()
```
